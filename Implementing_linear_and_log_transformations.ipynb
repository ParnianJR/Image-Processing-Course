```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img = Image.open('View.jpg')

plt.imshow(img, cmap='gray')
```

```python
img_array = np.asarray(img)
img_arr = img_array.flatten()
```

```python
#linear intensity transformation
#This is negative transformation:
alpha = float(input("Please enter alpha(linear intensity transformation): "))
beta = float(input("Please enter beta(linear intensity transformation): "))


img_list = list(img_array.flatten())

lin_img_list = []
for p in img_list:
    if alpha*p+beta > 255:
        lin_img_list.append(255)
    elif alpha*p+beta < 0:
        lin_img_list.append(0)
    else:
        lin_img_list.append(round(alpha*p+beta))

#lin_img_list = [round(alpha*p+beta) for p in img_list]

# reshape and write back into img_array
lin_img_array = np.reshape(np.asarray(lin_img_list), img_array.shape)
```

```python
fig = plt.figure()
fig.set_figheight(15)
fig.set_figwidth(15)

fig.add_subplot(1,2,1)
plt.imshow(img, cmap='gray')

# display the new image
fig.add_subplot(1,2,2)
plt.imshow(lin_img_array, cmap='gray')

plt.show(block=True)
```

```python
#log transformation
#setting c parameter when depth-length is 8
#c = 255/np.log2(256)
c = 255/np.log(256)
r = 1
#log_img_list = [c*np.log(p+r) for p in img_list]

# reshape and write back into img_array
#log_img_array = np.reshape(np.asarray(lin_img_list), img_array.shape)

#loga_img_array = np.rint(c * np.log(img_arr + 1))
# reshape and write back into img_array
log_img_array = np.reshape(np.rint(c * np.log(img_arr + 1)), img_array.shape)
log_img_array = np.array(log_img_array, dtype = np.uint8)
```

```python
fig = plt.figure()
fig.set_figheight(15)
fig.set_figwidth(15)

fig.add_subplot(1,2,1)
plt.imshow(img, cmap='gray')

# display the new image
fig.add_subplot(1,2,2)
plt.imshow(log_img_array, cmap='gray')

plt.show(block=True)
```
